#### Creating an evaluation matrix function

```python
def evaluation_matrix(actual_output_data, predict_output_data):
    print(f"f1 score: {f1_score(actual_output_data, predict_output_data,average='macro')}")
    print(f"Classification report: \n{classification_report(actual_output_data,predict_output_data)}")
#     print(f"Roc score: {roc_auc_score(actual_output_data, predictions, average='micro' ,multi_class='ovr')}")
```

### Performing the accuracy of various models on above dataset

```python
models={
    'Logistic Regression': LogisticRegression(),
    'KNN classifier': KNeighborsClassifier(),
    'naive bayes': GaussianNB(),
    'xgb': XGBClassifier()
#     'svm classifier': SVC(),
#     'Random Forest': RandomForestClassifier(class_weight='balanced'),
#     'Decision Tree': DecisionTreeClassifier(class_weight='balanced'),
    
}

x_train1=X
y_train1=Y
x_test1=X_TEST
y_test1=Y_TEST['disease']

high_time_consuming_models=['Logistic Regression', 'svm classifier']

params={
    'logsitic reg': {'penalty':['l1','l2', 'elasticnet'], 'C':[10,20,30],'multi_class':['auto','ovr', 'multinomial']},
    'knn classifier':{'n_neighbors':[3,4,5,6], 'weights':['distance'], 'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute']},
    'naive bayes':{},
    'xgb': {'booster': ['gbtree'], 'learning_rate':[0.1],'sampling_method':['uniform'],'colsample_bytree':[0.9], 
            'colsample_bylevel':[0.1], 'colsample_bynode':[0.1], 'scale_pos_weight':[2],
           'tree_method':['auto'], 'updater':['grow_colmaker'], 'num_parallel_tree':[1]}
#     'svc':{'C':[100,110,120], 'kernel':[ 'poly', 'rbf', 'sigmoid'], 'decision_function_shape':['ovo', 'ovr'], 'probability':[True]},
#     'random forest':{'criterion':['entropy', 'log_loss','gini'],'max_features':['sqrt', 'log2']},
#     'decision tree':{'criterion':['gini','entropy','log_loss'], 'max_features':['sqrt', 'log2'],'splitter':['best','random']}
    
}


for i in range(len(models)):
    
    gd_cv= GridSearchCV(models[list(models.keys())[i]], param_grid=params[list(params.keys())[i]], scoring='accuracy', cv=10)
    
    
    if list(models.keys())[i] in high_time_consuming_models:
        train=X
        y_train=Y
        test=X_TEST
        y_test=Y_TEST['disease']
        print(f"Dataset shape: {train.shape}")
        
        gd_cv.fit(train,y_train)
        
        print("Best Parameters: ",gd_cv.best_params_)
        training_predictions=gd_cv.predict(train)
        test_predictions=gd_cv.predict(test)
        
        print(f"Misclassified features: {(test_predictions!=y_test).sum()}")
        
        evaluation_matrix(y_train, training_predictions)
        evaluation_matrix(y_test, test_predictions)
        
    if list(models.keys())[i] not in high_time_consuming_models:
        train1=x_train1
        y_trainl=y_train1  # y_trainl --> larger training dataset
        test1=x_test1
        y_testl=y_test1
        print(f"Dataset shape: {train1.shape}")
        
        gd_cv.fit(train1,y_trainl)
        
        print("Best Parameters: ",gd_cv.best_params_)
        training_predictions=gd_cv.predict(train1)
        test_predictions=gd_cv.predict(test1)
        
        evaluation_matrix(y_trainl, training_predictions)
        print(f"Misclassified features test data: {(test_predictions!=y_testl).sum()}")
        evaluation_matrix(y_testl, test_predictions)
    
    
    print("------------------------------------------------------------------------------------------------------------------")
    
```

- Random Forest classifier accuracy --> 0.92
- KNN classifier accuracy --> 0.90
- SVM classifier accuracy --> 0.88
